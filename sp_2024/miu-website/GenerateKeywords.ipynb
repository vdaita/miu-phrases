```python
from gensim.models.keyedvectors import KeyedVectors
model = KeyedVectors.load_word2vec_format("GoogleNews-vectors-negative300.bin", binary=True)
```

```python
# seed_words_antiforeign = [
#     "competition", "Chinese", "communist", "foreign", "global", "international", "Japanese", "Korean", "multinational", "overseas", "trade war", "tariff"
# ]

seed_words_fairness = [
    "labor", "unions", "rights"
]
```

```python
import numpy as np
filtered_words = [word for word in seed_words_fairness if word in model.key_to_index]
print(f"Filtered {len(seed_words_fairness) - len(filtered_words)} words")
vectors = [model[word] for word in filtered_words]
average_vector = np.mean(vectors, axis=0)
print(average_vector)
```

```python
# Find 100 words close to this averaged vector.
similar_words = model.similar_by_vector(average_vector, topn=500)
similar_words = [word for word, _ in similar_words]
similar_words = [word.replace("_", " ") for word in similar_words]
similar_words = [word.lower() for word in similar_words]
print(similar_words)
print(len(similar_words))
```
